```python
import numpy
import matplotlib.pyplot as plt
import random
def fourier(q,l):
    return numpy.real(sum(map(lambda x:numpy.exp(-1j*q*x),l)))
l=[10000*(2*random.random()-1) for _ in range(100)]
q=[-2+0.02*x for x in range(200)]
fi=[fourier(qi,l) for qi in q]
plt.plot(q,fi)
plt.savefig("RPA.svg")
plt.show()
```
